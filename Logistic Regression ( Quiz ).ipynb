```python
# chi-square test of independence for independence between workclass and annual income:
import scipy.stats as stats
```

```python
cross=pd.crosstab(df['annual_income'],df['workclass']) # code for creating a cross-table.
```

```python
cross # Shundorr!!!
```

```python
stats.chi2_contingency(cross) # code for chi-square test of independence.
```

## Conclusion: Since here as per the chi-square test of independence we can see the p-value is wayyyy less than the level of sig (0.05) so we reject the null hypothesis that there exists independence between annual_inconme of an individual and his work_class. Thus work_class becomes an important variable in our model.
